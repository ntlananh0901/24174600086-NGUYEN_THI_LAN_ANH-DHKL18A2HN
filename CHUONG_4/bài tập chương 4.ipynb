```python
#4.1 Tạo chuỗi Pandas từ danh sách, numpy array và từ điển

import pandas as pd
import numpy as np

# Tạo dữ liệu
mylist = list('abcdefghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))

# Tạo Series từ các nguồn khác nhau
ser_from_list = pd.Series(mylist)
ser_from_array = pd.Series(myarr)
ser_from_dict = pd.Series(mydict)

print("Series từ list:\n", ser_from_list)
print("\nSeries từ array:\n", ser_from_array)
print("\nSeries từ dict:\n", ser_from_dict)
```

```python
#Bài 4.2: Chuyển chuỗi ser thành DataFrame với chỉ mục là một cột
# Tạo Series từ dict
ser = pd.Series(mydict)

# Chuyển thành DataFrame
df = ser.reset_index()
df.columns = ['index', 'value']

print("DataFrame từ Series:\n", df)
```

```python
#Bài 4.3: Kết nối hai Series thành một DataFrame
ser1 = pd.Series(list('abcdefghijklmnopqrstuvwxyz'))
ser2 = pd.Series(np.arange(26))

# Ghép thành DataFrame
df = pd.DataFrame({'letters': ser1, 'numbers': ser2})

print("DataFrame kết hợp:\n", df)
```

```python
#4.4 Thao tác với chỉ mục của Series và DataFrame
ser1 = pd.Series([1, 2, 3, 4, 5])

# Ví dụ: Tính tổng, trung bình, hoặc chuyển thành DataFrame
print("Tổng:", ser1.sum())
print("Trung bình:", ser1.mean())

# Chuyển thành DataFrame
df = ser1.to_frame(name='values')
print("DataFrame từ ser1:\n", df)
```

```python
#4.5Tính toán thống kê cơ bản từ Series
import pandas as pd
import numpy as np

ser = pd.Series(np.random.normal(10, 5, 25))

print("Giá trị tối thiểu:", ser.min())
print("Phần centile thứ 25:", ser.quantile(0.25))
print("Trung vị:", ser.median())
print("Phần centile thứ 75:", ser.quantile(0.75))
print("Giá trị tối đa:", ser.max())
```

```python
#4.6. Giải thích và đếm số lần xuất hiện
ser = pd.Series(np.take(list('abcdefghi'), np.random.randint(8, size=30)))

# Giải thích:
# np.random.randint(8, size=30): tạo 30 số ngẫu nhiên từ 0 đến 7
# np.take(...): lấy phần tử từ danh sách 'abcdefghi' theo chỉ số ngẫu nhiên

# Đếm số lần xuất hiện
counts = ser.value_counts()
print("Số lần xuất hiện của từng giá trị:\n", counts)
```

```python
#4.7Giữ lại 2 giá trị xuất hiện nhiều nhất, thay phần còn lại bằng 'Other'

np.random.seed(100)
ser = pd.Series(np.random.randint(1, 5, [12]))

# Tìm top 2 giá trị xuất hiện nhiều nhất
top2 = ser.value_counts().index[:2]

# Thay thế các giá trị khác bằng 'Other'
ser_modified = ser.where(ser.isin(top2), other='Other')
print("Series sau khi thay thế:\n", ser_modified)


```

```python
#Bài 4.8: Chia thành 10 phân vị bằng nhau và gán tên phân vị
ser = pd.Series(np.random.random(20))

# Chia thành 10 phân vị
labels = [f'Decile {i+1}' for i in range(10)]
ser_deciled = pd.qcut(ser, q=10, labels=labels)

print("Series sau khi gán phân vị:\n", ser_deciled)
```

```python
#2. Định hình lại ser thành một DataFrame có 7 hàng và 5 cột:
import pandas as pd
import numpy as np

# Tạo Series
ser = pd.Series(np.random.randint(1, 10, 35))

# Chuyển thành DataFrame 7 hàng x 5 cột
df = pd.DataFrame(ser.values.reshape(7, 5), columns=[f'col{i+1}' for i in range(5)])

print("DataFrame 7x5:\n", df)
```
