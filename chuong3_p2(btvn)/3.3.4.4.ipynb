```python
import numpy as np
arr = np.arange(16).reshape((4,4))
arr
```

```python
upper, lower =np.vsplit(arr,2) # tách theo trục ngang
print("Kết quả sau khi tách mảng từ ban đầu ")
print(upper)
print(lower)
```

```python
left, right=np.hsplit(arr,2) # tách theo trục dọc
print("Kết quả sau khi tách từ mảng ban đầu")
print(left)
print(right)
```
