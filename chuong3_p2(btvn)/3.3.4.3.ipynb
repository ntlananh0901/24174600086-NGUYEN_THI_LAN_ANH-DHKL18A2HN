```python
import numpy as np
arr = np.arange(10)# tạo mảng gồm 10 phần tử
sub_arr1=np.split(arr,[3,6]) # chia mảng arr theo chỉ mục [3,6]
#Phần đầu tiên bắt đầu từ mục 0 và kết thúc trc chỉ mục 3
# Phần thứ 2 bắt đầu từ chỉ mục 3 và kết thúc trc chỉ mục 6

sub_arr2=np.split(arr,2) # chia mảng thành 2 phần bằng nhau
print("Mảng con sub_arr1 là: ", sub_arr1 )
print("Mảng con sub_arr2 là", sub_arr2)
```
